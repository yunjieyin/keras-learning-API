```python
# import relative libs

import warnings
warnings.filterwarnings('ignore')
import os
import matplotlib.pyplot as plt
```

### Data and image path

```python
# current working directory
ROOT_DIR = os.getcwd()

# data directory
DATA_PATH = os.path.join(ROOT_DIR, 'data')

# image path
img_path = os.path.join(DATA_PATH, 'car.jpg')
```

### Use pillow lib process image

```python
from PIL import Image

# read image
image = Image.open(img_path)

# convert image's format and save
image.save(os.path.join(DATA_PATH, 'car_cp.png'))

#Display image
plt.imshow(image)
plt.show()
```

### Resize image

```python
image = Image.open(img_path)

# resize
img_resized = image.resize((400, 400))

print('origin image size: ', image.size)
print('resized image size: ', img_resized.size)

plt.imshow(img_resized)
plt.show()
```

### Resize keep width and height ratio

```python
image = Image.open(img_path)
image.thumbnail((200, 200))

print("new image size: ", image.size)

plt.imshow(image)
plt.show()
```

### Cropping

```python
image = Image.open(img_path)

#cropping bbox
x1 = 100
y1 = 50
x2 = 200
y2 = 150
bbox = (x1, y1, x2, y2)

img_cropped = image.crop(bbox)

plt.imshow(img_cropped)
plt.show()
```

### Rotating

```python
# keep origin image's size

image = Image.open(img_path)

img_rot_90 = image.rotate(90)

plt.imshow(img_rot_90)
plt.show()

img_rot_180 = image.rotate(180)

plt.imshow(img_rot_180)
plt.show()
```

```python
# rotate image with expand operation

img = Image.open(img_path)

img_rot_18_expand = img.rotate(18, expand=True)

plt.imshow(img_rot_18_expand)
plt.show()
```

### Flipping

```python
img = Image.open(img_path)

img_flip = img.transpose(Image.FLIP_LEFT_RIGHT)

plt.imshow(img_flip)
plt.show()

img_flip = img.transpose(Image.FLIP_TOP_BOTTOM)

plt.imshow(img_flip)
plt.show()
```

### Drawing

```python
from PIL import Image, ImageDraw, ImageFont

# create a blank image
img_blank = Image.new('RGBA', (400, 300), 'white')
img_draw = ImageDraw.Draw(img_blank)

# draw a rectangle
img_draw.rectangle((70, 50, 270, 200), outline='red', fill='blue')

# get a font object
#fnt = ImageFont.truetype('comic.ttf', 40)

# draw text
img_draw.text((70, 250), 'Hello World', fill='green')

# show
plt.imshow(img_blank)
plt.show()
```

```python
img_blank = Image.new('RGBA', (400, 300), 'white')
img_draw = ImageDraw.Draw(img_blank)

img_draw.rectangle((70, 50, 270, 200), outline=None, fill='pink')

img_draw.line([(70, 50), (270, 50), (270, 200), (70, 200), (70, 50)], fill='red', width=4)

r = 10
img_draw.ellipse((270 - r, 200 - r, 270 + r, 200 + r), fill='orange')

img_draw.polygon([(40, 40), (40, 80), (80, 60), (60, 40)], fill='green', outline=None)

plt.imshow(img_blank)
plt.show()
```

### Convert Color

```python
img = Image.open(img_path)

img_grey = img.convert('L')

plt.imshow(img_grey)
plt.show()
```
